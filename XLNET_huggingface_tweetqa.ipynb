```python
xlnet_finetuned_squad_answers = []
qa_model = pipeline('question-answering', tokenizer='ahotrod/roberta_large_squad2', model='ahotrod/roberta_large_squad2')
i = 0
for index, row in test.iterrows():
    ans = qa_model({'question': row['Question'], 'context': row['Tweet']})['answer']
    xlnet_finetuned_squad_answers.append(ans)
    if i%10 == 0:
        print(i,'Done')
    i+=1
```
